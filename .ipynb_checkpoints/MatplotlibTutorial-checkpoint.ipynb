```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

plt.style.use('default') #can change this to a seaborn or ggplot
```

```python
x = range(7)
y = [4,8,12,16,20,24,28]

plt.plot(x,y)
plt.plot(y,x)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Plot')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Line Plot (Object-Oriented)')
plt.show()
```

# Basic Line Plots


```python
x = np.linspace(0,10,100) # 100 points from 0 to 10
y = np.sin(x)

plt.figure(figsize = (8, 5))
plt.plot(x,y)
plt.xlabel('X values')
plt.ylabel('sin(X)')
plt.title('Sin Wave')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0,10,100) # 100 points from 0 to 10
y = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize = (10, 6))
plt.plot(x,y, label='(sin of x)')
plt.plot(x,y2, label='(cos of x)')
plt.plot(x,y3, label='(tan of x)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Trigonometric Waves')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

# Customizing Plots: Colors, Styles, and Markers

### Matpolotlib offers extensive customization options to make your plots visually appealing and informative.

```python
x = np.linspace(0,10,20)

# Different line styles and markers
plt.figure(figsize = (12,6))

plt.plot(x,x,'b-', label='Solid line', linewidth=2)
plt.plot(x,x+1,'r--', label='Dashed line', linewidth=2)
plt.plot(x,x+2,'g-', label='Dash-dot line', linewidth=2)
plt.plot(x,x+3,'m:', label='Dotted line', linewidth=2)
plt.plot(x,x+4,'co-', label='Circles', markersize=8)
plt.plot(x,x+5,'s-', label='Squares', markersize=6)
plt.plot(x,x+6,'^-', label='Triangles', markersize=6)

plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Line and styles')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Grouped bar chart

categories = ['1','2','3','4']
product_a =[20,43,12,32]
product_b =[10,33,42,22]
product_c =[29,24,42,2]

x = np.arange(len(categories))
width = 0.25

plt.figure(figsize=(10,6))

plt.bar(x - width, product_a,  width, label='A',color='blue')
plt.bar(x , product_b, width, label='B', color='red')
plt.bar(x + width, product_c, width, label='C', color='yellow')

plt.xlabel('Quarter')
plt.ylabel('sales')
plt.title('Quarterly Sales Comparison')
plt.xticks(x,categories)
plt.legend()
plt.grid(True, alpha =0.3, axis='y')
plt.show()
```

# Histograms

```python
np.random.seed(42)
data =  np.random.normal(100,15,1000)

# Basic histogram
plt.figure(figsize=(10,6))
plt.hist(data, bins=30, color='skyblue',edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frecuency')
plt.title('Histogram of ND')
plt.grid(True, alpha=0.3, axis'y')
plt.show()
```
