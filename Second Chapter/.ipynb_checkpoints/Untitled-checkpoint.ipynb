```python
boston_re = boston.set_index('Unnamed: 0')
```

```python
boston_re
```

```python
from matplotlib.pyplot import subplots
fig, axes = subplots(ncols = 3,nrows=3, figsize = (15,15))
boston_re.plot.scatter('crim','nox', ax = axes[0][0])
boston_re.plot.scatter('crim','dis', ax = axes[1][0])
boston_re.plot.scatter('crim','indus', ax = axes[2][0])
boston_re.plot.scatter('tax','nox', ax = axes[0][1])
boston_re.plot.scatter('tax','dis', ax = axes[1][1])
boston_re.plot.scatter('tax','indus', ax = axes[2][1])
boston_re.plot.scatter('ptratio','nox', ax = axes[0][2])
boston_re.plot.scatter('ptratio','dis', ax = axes[1][2])
boston_re.plot.scatter('ptratio','indus', ax = axes[2][2])
```
